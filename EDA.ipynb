### 2. Data Preparation

```python
# Choose relevant columns
df.columns
```

```python
df = df[[#'uuid', 'City',
    'Surface_area', 'Potential_installable_area',
       'Peak_installable_capacity', 'Energy_potential_per_year',
       'Assumed_building_type', 'Estimated_tilt', 
        #'Unit_installation_price', 'Comment', 'Estimated_capacity_factor',
       'Estimated_building_height',]].copy()
```

```python
# Rename our columns
df.columns
```

```python
df = df.rename(columns = {'Surface_area':'Surface_Area',
                     'Potential_installable_area': 'Potential_Installable_Area',
                     'Peak_installable_capacity':'Peak_Installable_Capacity',
                     'Energy_potential_per_year':'Energy_Potential_per_Year',
                     'Assumed_building_type':'Building_Type',
                     'Estimated_tilt':'Estimated_Tilt',
                      'Estimated_building_height':'Estimated_Building_Height',
                      'Estimated_capacity_factor':'Estimated_Capacity_Factor' }).copy()
```

```python
# Creation of 'percent capacity factor' and 'Estimated_building_volume' columns
df['Estimated_Capacity_Factor%'] = (df['Energy_Potential_per_Year'] / (df['Peak_Installable_Capacity'] * 8760)) * 100
df['Estimated_Building_Volume'] = (df['Surface_Area'] * df['Estimated_Building_Height']) / 10**6
```

```python
df.head()
```

```python
df.loc[df.duplicated()]
```

```python
df = df.loc[~df.duplicated()] 
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Surface area < 2 metre square is unrealistic.
df = df.query('Surface_Area > 2')

# Building less than 10 feet -- 3.05 metres is unrealistic!
df = df.query('Estimated_Building_Height > 3.05')
```

```python
df['Estimated_Building_Height'].fillna(df['Estimated_Building_Height'].mean(), inplace = True)
df['Estimated_Capacity_Factor%'].fillna(df['Estimated_Capacity_Factor%'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

#### Question: What is the total estimated solar energy generated by each building type per year?

```python
 df.groupby('Building_Type')['Energy_Potential_per_Year']\
.agg(['mean', 'sum', 'count']).sort_values(by = 'sum', ascending = False)
```

```python
ax = df.groupby('Building_Type')['Energy_Potential_per_Year']\
.agg(['mean', 'sum', 'count']).sort_values(by = 'sum')['sum']\
.plot(kind = 'barh', figsize = (12,5), title = 'Sum of Estimated Solar Energy Potential by Building Type')
ax.set_xlabel('Sum of Estimated Solar Energy Poential [GWh]')
ax.set_ylabel('Building Type')

ax.set_xticklabels([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000])


plt.show()
```
